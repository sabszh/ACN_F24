```python
import pandas as pd
import glob
import os
```

```python
file_path_A = "../../../../MEG_data/logfiles/145A*.csv"
file_path_B = "../../../../MEG_data/logfiles/145B*.csv"
```

```python
all_files_A = glob.glob(file_path_A)
all_files_B = glob.glob(file_path_B)
```

```python
df_A = pd.concat((pd.read_csv(f) for f in all_files_A), ignore_index=True)
df_B = pd.concat((pd.read_csv(f) for f in all_files_B), ignore_index=True)
```

```python
df_A["section"] = "A"
df_B["section"] = "B"
```

```python
df_combined = pd.concat([df_A, df_B], ignore_index=True)
```

```python
df_A.to_csv('../in/merged_logfile_145_A.csv', index=False)
df_B.to_csv('../in/merged_logfile_145_B.csv', index=False)

df_combined.to_csv('../in/merged_logfile_145_all.csv', index=False)
```
